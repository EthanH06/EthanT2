<script>
    const BITS = {{ BITS }};
    const MAX = 2 ** BITS - 1;
    const MSG_ON = "on";
    const IMAGE_ON = "{{site.baseurl}}/images/bulb_on.gif";
    const MSG_OFF = "off";
    const IMAGE_OFF = "{{site.baseurl}}/images/bulb_off.png";
    let slots = [0, 0, 0, 0, 0, 0, 0, 0];
    let totalValue = 0;
    // Function to get binary input from user and update bits
    function getUserInputAndUpdateBits() {
        for (let i = 0; i < BITS; i++) {
            let binaryDigit = prompt(`Insert a binary digit (0 or 1) for position ${i + 1}: `);
            if (binaryDigit === '0') {
                slots[i] = 0;
            } else if (binaryDigit === '1') {
                slots[i] = Math.pow(2, BITS - 1 - i);
            } else {
                alert("Invalid input. Please enter either 0 or 1.");
                throw new Error("Invalid input.");
            }
            // Update the corresponding DOM elements based on user input
            updateBits();
        }
    }
    // Function to update bits and visual representation based on user input
    function updateBits() {
        for (let i = 0; i < BITS; i++) {
            let digit = slots[i];
            document.getElementById('digit' + i).value = digit;
            const image = document.getElementById('bulb' + i);
            const butt = document.getElementById('butt' + i);
            if (digit === 1) {
                image.src = IMAGE_ON;
                butt.innerHTML = MSG_OFF;
            } else {
                image.src = IMAGE_OFF;
                butt.innerHTML = MSG_ON;
            }
        }
    }
    // Function to handle bit toggling
    function toggleBit(i) {
        const image = document.getElementById('bulb' + i);
        const butt = document.getElementById('butt' + i);
        // Toggle bit value
        slots[i] = slots[i] === 0 ? 1 : 0;
        // Change visual representation
        if (slots[i] === 0) {
            image.src = IMAGE_OFF;
            butt.innerHTML = MSG_ON;
        } else {
            image.src = IMAGE_ON;
            butt.innerHTML = MSG_OFF;
        }
        // Recalculate total value
        totalValue = slots.reduce((sum, value) => sum + value, 0);
        // Print the total value
        if (totalValue > MAX) {
            console.log("Total value: ff");
        } else {
            console.log("Total value:", totalValue);
        }
    }
    // ... (rest of your existing code)
</script>
